# **sentiment Analysis**

```python
# function to print sentiments
# of the sentence.
def sentiment_scores(sentence):
# Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
    sentiment = None
# polarity_scores method of SentimentIntensityAnalyzer
# oject gives a sentiment dictionary.
# which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
#     print("Overall sentiment dictionary is : ", sentiment_dict)
#     print("sentence was rated as ", sentiment_dict['neg']*100, "% Negative")
#     print("sentence was rated as ", sentiment_dict['neu']*100, "% Neutral")
#     print("sentence was rated as ", sentiment_dict['pos']*100, "% Positive")
#     print("Sentence Overall Rated As", end = " ")
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        #print("Positive")
        sentiment = 'Positive'
    elif sentiment_dict['compound'] <= - 0.05 :
        #print("Negative")
        sentiment = 'Negative'
    else :
        #print("Neutral")
        sentiment = 'Neutral'
    return sentiment
```

```python
df['sentiment'] = df['text'].apply(sentiment_scores)
```

# **preprocessing**

```python
df
```

```python
import re
lemma_ = WordNetLemmatizer()
def clean_text(text):
    text = str(text)
    text = " ".join(text.split())  
    text = re.sub('[^a-zA-Z]', " ", text) #remove punctuations and numbers
    text = re.sub(r"\s+[a-zA-Z]\s+", ' ', text) # Single character removal
    text = re.sub('(@[A-Za-z]+[A-Za-z0-9-_]+)', '', text) #remove_user
    text = re.sub(r'http\S+', '', text)   #remove http liks
    text = re.sub(r'bit.ly/\S+', '', text)  #remove other web links
    text = text.strip('[link]')
    html = re.compile('<.*?>')
    text =  html.sub(r'', text)
    email = re.compile(r'[\w\.-]+@[\w\.-]+')
    text = email.sub(r'',text)
    text = re.sub(r'\s+', " ", text) #remove extra spaces
    text = text.replace("ain't", "am not").replace("aren't", "are not")
    text = ' '.join(tex.lower() for tex in text.split(' ')) # Lowering cases
    sw = stopwords.words('english')
    text = ' '.join(tex for tex in text.split() if tex not in sw) #removing stopwords
    text = ' '.join(lemma_.lemmatize(x) for x in text.split()) #lemmatization
    return text
```

```python
df['clean_text'] = df['text'].apply(clean_text)
```

```python
df['clean_text']
```

# **TFIDF Vectorization**

```python
vec = TfidfVectorizer(max_features=5000)
```

```python
X = df['clean_text']
y = df['sentiment']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
train_vectors = vec.fit_transform(X_train)
test_vectors = vec.transform(X_test)
```

```python
train_vectors = train_vectors.toarray()
test_vectors = test_vectors.toarray()
```

```python
y_test
```

```python
y_test = y_test.apply(lambda x: 0 if x=='Positive' else (2 if x=='Negative' else 1))
y_train = y_train.apply(lambda x: 0 if x=='Positive' else (2 if x=='Negative' else 1))
y_train = y_train.tolist()
y_test = y_test.tolist()
y_train, y_test
```

# **Modeling**

## **Random Forest model**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(train_vectors, y_train)
```

```python
y_pred = rf_model.predict(test_vectors)
print(classification_report(y_test, y_pred))
```

## **SVM**

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC()
svm_model.fit(train_vectors, y_train)
```

```python
y_pred = svm_model.predict(test_vectors)
print(classification_report(y_test, y_pred))
```

```python
lr_model = LogisticRegression()
lr_model.fit(train_vectors, y_train)
```

```python
import numpy
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
# fix random seed for reproducibility
numpy.random.seed(7)
```

```python
top_words = 5000
from keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(num_words=top_words)
```

```python
tokenizer.fit_on_texts(X_train)
X_train = tokenizer.texts_to_sequences(X_train)
X_test = tokenizer.texts_to_sequences(X_test)
```

```python
max_review_length = 500
X_train = sequence.pad_sequences(X_train, maxlen=max_review_length)
X_test = sequence.pad_sequences(X_test, maxlen=max_review_length)
```

```python
embedding_vecor_length = 32
model = Sequential()
model.add(Embedding(top_words, embedding_vecor_length, input_length=max_review_length))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=64)
```
